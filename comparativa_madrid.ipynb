# Real madrid 25-26 vs Real madrid 23-24

```python
import pandas as pd
```

```python
team_25_26 = pd.read_csv("RM2526/team-players.csv",sep=";")
situacion_25_26 = pd.read_csv("RM2526/team-statistics.csv",sep=";")
formacion_25_26 = pd.read_csv("RM2526/team-statistics (1).csv",sep=";")
game_estate_25_26 = pd.read_csv("RM2526/team-statistics (2).csv",sep=";")
timing_25_26 = pd.read_csv("RM2526/team-statistics (3).csv",sep=";")
shot_zones_25_26 = pd.read_csv("RM2526/team-statistics (4).csv",sep=";")
attack_speed_25_26 = pd.read_csv("RM2526/team-statistics (5).csv",sep=";")
result_25_26 = pd.read_csv("RM2526/team-statistics (6).csv",sep=";")
```

```python
team_23_24 = pd.read_csv("RM2324/team-players (1).csv",sep=";")
situacion_23_24 = pd.read_csv("RM2324/team-statistics (7).csv",sep=";")
formacion_23_24 = pd.read_csv("RM2324/team-statistics (8).csv",sep=";")
game_estate_23_24 = pd.read_csv("RM2324/team-statistics (9).csv",sep=";")
timing_23_24 = pd.read_csv("RM2324/team-statistics (10).csv",sep=";")
shot_zones_23_24 = pd.read_csv("RM2324/team-statistics (11).csv",sep=";")
attack_speed_23_24 = pd.read_csv("RM2324/team-statistics (12).csv",sep=";")
result_23_24 = pd.read_csv("RM2324/team-statistics (13).csv",sep=";")
team_24_25 = pd.read_csv("team-players (2).csv",sep=";")
```

```python
formacion_25_26.rename(columns={"statistic": "formacion"}, inplace=True)
situacion_25_26.rename(columns={"statistic": "situacion"}, inplace=True)
game_estate_25_26.rename(columns={"statistic": "game_estate"}, inplace=True)
timing_25_26.rename(columns={"statistic": "timing"}, inplace=True)
shot_zones_25_26.rename(columns={"statistic": "shot_zones"}, inplace=True)
attack_speed_25_26.rename(columns={"statistic": "attack_speed"}, inplace=True)
result_25_26.rename(columns={"statistic": "result"}, inplace=True)
formacion_23_24.rename(columns={"statistic": "formacion"}, inplace=True)
situacion_23_24.rename(columns={"statistic": "situacion"}, inplace=True)
game_estate_23_24.rename(columns={"statistic": "game_estate"}, inplace=True)
timing_23_24.rename(columns={"statistic": "timing"}, inplace=True)
shot_zones_23_24.rename(columns={"statistic": "shot_zones"}, inplace=True)
attack_speed_23_24.rename(columns={"statistic": "attack_speed"}, inplace=True)
result_23_24.rename(columns={"statistic": "result"}, inplace=True)
```

```python
df_2526=[team_25_26,formacion_25_26,situacion_25_26,game_estate_25_26,timing_25_26,shot_zones_25_26,attack_speed_25_26,result_25_26]
df_2324=[team_23_24,formacion_23_24,situacion_23_24,game_estate_23_24,timing_23_24,shot_zones_23_24,attack_speed_23_24,result_23_24]
for df in df_2526:
    df["season"] = "25-26"
for df in df_2324:
    df["season"] = "23-24"
team_24_25["season"] = "24-25"
```

```python
formacion=pd.concat([formacion_25_26, formacion_23_24], ignore_index=True)
situacion=pd.concat([situacion_25_26, situacion_23_24], ignore_index=True)
game_estate=pd.concat([game_estate_25_26, game_estate_23_24], ignore_index=True)
timing=pd.concat([timing_25_26, timing_23_24], ignore_index=True)
shot_zones=pd.concat([shot_zones_25_26, shot_zones_23_24], ignore_index=True)
attack_speed=pd.concat([attack_speed_25_26, attack_speed_23_24], ignore_index=True)
result=pd.concat([result_25_26, result_23_24], ignore_index=True)
team=pd.concat([team_25_26, team_23_24, team_24_25], ignore_index=True)
```

```python
dfs= [formacion, situacion, game_estate, timing, shot_zones, attack_speed, result, team]
for df in dfs:
    df.drop(columns=['number'], inplace=True)
```

# team

```python
team["linea"] = team["position"].apply(
    lambda x: "Ataque" if "F" in x
    else "Medio" if "M" in x
    else "Portero" if "GK" in x
    else "Defensa"
)
```

```python
team.sort_values(by="linea", ascending=False)
```

```python
team.loc[team["player"]=="Franco Mastantuono", "linea"] = "Ataque"
team.loc[team["player"]=="Arda Güler", "linea"] = "Medio"
team.loc[team["player"]=="Jude Bellingham", "linea"] = "Medio"
team.loc[team["player"]=="Federico Valverde", "linea"] = "Medio"
```
